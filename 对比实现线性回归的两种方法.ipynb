 ## 这是一个用梯度下降算法实现的线性回归。

```python
import numpy as np
from matplotlib import pyplot as plt

def loss_function(k, d, X, Y):
    length = len(X)
    loss = 0
    for i in range(0, length):
        loss += (Y[i]-(k*X[i]+d)) ** 2
    return loss
 
def gradient_descent(X, Y, learning_rate, epochs):
    # 初始化斜率和截距
    k = 0
    d = 0
    N = len(X)
    
    # 进行迭代更新参数
    for _ in range(epochs):
        # 计算损失函数关于斜率和截距的偏导数的平均值
        dk = (-2/N) * sum(X * (Y - (k*X + d)))
        dd = (-2/N) * sum(Y - (k*X + d))
        
        # 更新参数
        k = k - learning_rate * dk
        d = d - learning_rate * dd
    
    return k, d
    
```

```python
np.random.seed(123)
X_1 = np.arange(1, 21)
Y_1 = 2.3 * X_1 + np.random.uniform(-5, 5, 20)

plt.scatter(X_1, Y_1)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of X vs Y')
plt.show()
```

```python
# 设置学习率和迭代次数
learning_rate = 0.0001
epochs = 10000

# 调用梯度下降函数进行线性拟合
k, d = gradient_descent(X_1, Y_1, learning_rate, epochs)

print("斜率 k:", k)
print("截距 d:", d)
```

```python
loss_function(k, d, X_1, Y_1)
```

```python
X_1_predict = np.random.uniform(0, 20, size=(1000,))
Y_1_predict = X_1_predict * k + d
plt.plot(X_1_predict, Y_1_predict, color='red')
plt.scatter(X_1, Y_1)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Plot of Predicted X vs Y')
plt.show()
```

## 用numpy的polyfit函数拟合。

```python
import numpy as np
from matplotlib import pyplot as plt

# 进行线性回归
slope, intercept = np.polyfit(X_1, Y_1, 1)
k = slope
d = intercept
print("k="+str(k)+", d="+str(d))
# 绘制散点图
plt.scatter(X_1, Y_1, label='Data Points')

# 绘制拟合直线
plt.plot(X_1, slope*X_1 + intercept, color='red', label='Fitted Line')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression of X vs Y')
plt.legend()
plt.show()
```
